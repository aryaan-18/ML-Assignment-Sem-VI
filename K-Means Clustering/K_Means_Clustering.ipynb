```python
#Storing All Values Of Certain Columns In Independent Variable
x = data.iloc[:, [2,3]].values
print(x[0])
print(x[1])
```

```python
#Importing The KMeans Module From The Sklearn Library
from sklearn.cluster import KMeans
list1 = []

#Iterating Over A Range OF Clusters From 1 to 10
for i in range(1,11):
  #Initializing The K-Means Clustering Algorithm With Current Cluster Number
  K_means = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
  #Training The KMeans Model With The Data
  K_means.fit(x)
  #Appending The Inertia Value To The List (Sum Of Squared Distances
  #Of The Samples To Their Closest Cluster Center)
  list1.append(K_means.inertia_)
  print(list1)

#Plotting The Elbow Method Graph For Number Of Clusters And Inertia Values For Each Cluster Number
plt.plot(range(1, 11), list1)
plt.title('Elbow Method Graph')
plt.xlabel('Number Of Clusters')
plt.ylabel('List1 Output')
plt.show()
```

```python
#Initializing The K-Means Clustering Algorithm With Specified Parameters
k_means = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)

#Training The KMeans Model On The Data
y_k_means = k_means.fit_predict(x)

#Print The Data Points Corresponding To Each Cluster
print(x[y_k_means==0, 0], x[y_k_means==0, 1] )
print(x[y_k_means==1, 0], x[y_k_means==1, 1] )
print(x[y_k_means==2, 0], x[y_k_means==2, 1] )
```

```python
#Initializing The K-Means Clustering Algorithm With Specific Parameters
k_means1 = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 200, n_init = 10, random_state = 0)

#Training The K-Means Model On The Data
y_k_means1 = k_means1.fit(x)
```

```python
#Importing The Metrics And Pairwise_Distances Module From The Sklearn Library
from sklearn import metrics
from sklearn.metrics import pairwise_distances

#Accessing Cluster Labels And Displaying It
labels = y_k_means1.labels_
print(labels)

#Computing The Silhouette Score
metrics.silhouette_score(x, labels, metric = 'euclidean')
```
